```python
# Import NumPy
import numpy as np

# Create a NumPy array from a list
my_list = [1, 2, 3, 4, 5]
my_array = np.array(my_list)
print("My array:", my_array)
print(type(my_array))
```

```python
# Create a NumPy array of zeros
my_zeros = np.zeros(5)
print("Zeros:", my_zeros)
```

```python
# Create a NumPy array of ones
my_ones = np.ones(5)
print("Ones:", my_ones)
```

```python
# Create two NumPy array with a range of values, one is odd number from 0 to 10, another one is even number from 0 to 10
odd_numbers = np.arange(1, 10, 2)
even_numbers = np.arange(2, 11, 2)
print("Odds:", odd_numbers)
print("Evens:", even_numbers)
```

```python
# Reshape a NumPy array of 1(row)x6(column) to 2(row)x3(column)
even_numbers_with_zero = np.arange(0,11,2)
my_reshape = np.reshape(even_numbers_with_zero, (2, 3))
print("Reshaped array:\n", my_reshape)
```

```python
# Indexing and slicing a NumPy array

# Get the first element 0 from the even array
print("First element:", my_reshape[0, 0])

# Get all elements in second row
print("Second row:", my_reshape[1, :])

# Get only the last column
print("Last column:", my_reshape[:, -1])
```

```python
# Basic array operations
my_array1 = np.array([1, 2, 3])
my_array2 = np.array([4, 5, 6])

# Adding elements between two rows, i.e. [array1_column1+array2_column1, array1_column1+array2_column1, array1_column1+array2_column1]
print("Addition:", my_array1 + my_array2)

# Substracting elements in two rows, i.e. [array1_column1-array2_column1, array1_column1-array2_column1, array1_column1-array2_column1]
print("Subtraction:", my_array1 - my_array2)

# Multiplication elements in two rows, i.e. [array1_column1 x array2_column1, array1_column1 x array2_column1, array1_column1 x array2_column1]
print("Multiplication:", my_array1 * my_array2)

# Division elements in two rows, i.e. [array1_column1 / array2_column1, array1_column1 / array2_column1, array1_column1 / array2_column1]
print("Division:", my_array1 / my_array2)
```
